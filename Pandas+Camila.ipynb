# Exercício 1
Com este arquivo, construir um objeto de Serie cronológica com o índice igual ao ano dos jogos olímpicos

```python
import pandas as pd
import numpy as np
```

```python
om = pd.read_csv('bases/Summer-Olympic-medals-1976-to-2008.csv', encoding='ISO-8859-1')
om.head(3)
```

```python
om.info()
```

```python
om = om.dropna()
om['Year'] = om['Year'].astype('int').astype(str)
om.head(3)
```

```python
om.shape
```

```python
df = om.set_index('Year')
df.head(3)
```

# Exercício 2
Separar os nomes próprios de sobrenomes em duas colunas distintas usando vírgula como separador.

```python
df.head(3)
```

```python
nome_atleta = df['Athlete']
nome_atleta
```

```python
df[['Last_name','First_name']] = nome_atleta.str.rsplit(",", expand = True, n=1)
df.head(3)
```
